```python
#EC. DE MOVIMIENTO DE TIRO PARABÓLICO
#Autor: Flor Marián Galeana Juárez
#Fecha 1 de octubre del 2021
import matplotlib.pyplot as plt
import numpy as np
#Definimos funciones:
def x(t):
  return float(r1x)+float(v1x)*t #-0
def x1(t):
  return (x(t+intervalo)-x(t))/intervalo
def x2(t):
  return (x1(t+intervalo)-x1(t))/intervalo
def y(t):
  return float(r1y)+float(v1y)*t-(1/2)*g*t**2
def y1(t):
  return (y(t+intervalo)-y(t))/intervalo
def y2(t):
  return (y1(t+intervalo)-y1(t))/intervalo
#Lectura de variables:
g= 9.81 #m/s**2 constante
t= input("¿En qué tiempo quiere calcular el movimiento del proyectil? ")
intervalo= 0.001
r1x= input("Inserte la componente horizontal de la posición inicial del proyectil en metros: ")
r1y= input("Inserte la componente vertical de la posición inicial del proyectil en metros: ")
v1x= input("Inserte la componente horizontal de la velocidad inicial del proyectil en m/s: ")
v1y= input("Inserte la componente vertical de la velocidad inicial del proyectil en m/s: ")
t1= np.arange(0, 50, 0.01)
#Calculamos magnitudes de los vectores:
MVectorPosicion= np.sqrt((x(float (t))**2)+(y(float (t))**2))
MVectorVelocidad= np.sqrt((x1(float (t))**2)+(y1(float (t))**2))
MVectorAceleracion= np.sqrt((x2(float (t))**2)+(y2(float (t))**2))
#Imprimimos resultados:
print("A los", t, "segundos obtenemos los siguientes resultados sobre el movimiento del proyectil: ")
print("Las componentes del  vector posición son: (", (x(float (t))**2), "i,", (y(float (t))**2), "j) m")
print("La magnitud del vector posición es: ", MVectorPosicion, "metros.")
print("Las componentes del  vector velocidad son: (", (x1(float (t))**2), "i,", (y1(float (t))**2), "j) m/s")
print("La magnitud del vector velocidad es: ", MVectorVelocidad, "m/s.")
print("Las componentes del  vector aceleración son: (", (x2(float (t))**2), "i,", (y2(float (t))**2), "j) m/s**2")
print("La magnitud del vector aceleración es: ", MVectorAceleracion, "m/s**2.")
#Graficamos funciones:
plt.plot(t1, x(t1))
plt.plot(t1, y(t1))
plt.title("gráfica del vector posición")
plt.show()
plt.plot(t1, x1(t1))
plt.plot(t1, y1(t1))
plt.title("gráfica del vector velocidad")
plt.show()
plt.plot(t1, x2(t1))
plt.plot(t1, y2(t1))
plt.title("gráfica del vector aceleración")
plt.show()
```
