```python
#Nombre del programa: Programa para probar el round-error del Tiro parabólico con el método de euler
#Autor: Joseph Antonio Garnica Romero
#Fecha de creación: 25/10/2021

import numpy as np
import matplotlib.pyplot as plt



#Función para la velocidad:

def Velocidad(Vi,H,A,Nintervalos):
  suma=Vi;
  for i in range(1,Nintervalos):
    Velocidad= H*A
    suma=suma + Velocidad
  return suma



#Función para la posición:

def Posicion(Pi,H,Vi,A,Nintervalos):
  suma=Pi;
  Velocidad= Vi;
  for i in range(1,Nintervalos):
    Velocidad=Velocidad + (H*A)
    Posicion= H*Velocidad
    suma=suma + Posicion
  return suma

#Funciones para optener la velocidad exacta
def Velocidadx():
  return float(VXi)

def Velocidady():
  return float(VYi)-AY*T

#Funciones para optener la posición exacta
def Posicionx():
  return float(Xi)+(VXi*T)

def Posiciony():
  return float(Yi+(VYi*T)-(.5*AY*(T**2)))

#Inicializador de variables

print ("Incerte el valor de la posicion inicial en la componente x ")
Xi = float (input(""))

print ("Incerte el valor de la posicion inicial en la componente y ")
Yi = float (input(""))

print ("Incerte el valor de la velocidad inicial en la componente x ")
VXi = float (input(""))

print ("Incerte el valor de la velocidad inicial en la componente y ")
VYi = float (input(""))


print ("Incerte el valor del tiempo ")
T= float (input(""))

#Valor de la aceleración en X
AX=0

#Valor de la aceleración en Y
AY=-9.8



#Operaciones con distintos tamaños de intrervalos

vector_velocidadx=[]
vector_velocidady=[]
vector_posicionx=[]
vector_posiciony=[]
vector_velocidadxdeltah=[]
vector_velocidadydeltah=[]
vector_posicionxdeltah=[]
vector_posicionydeltah=[]

hrange=np.arange(-8,-1,1,dtype=np.float)

for hi in hrange:
  Ni= int(T/(10.0**hi))
  VX=Velocidad(VXi,10.0**hi,AX,Ni)
  vector_velocidadx.append(VX)
  vector_velocidadxdeltah.append(Velocidadx()-VX)
  VY=Velocidad(VYi,10.0**hi,AY,Ni)
  vector_velocidady.append(VY)
  vector_velocidadydeltah.append(Velocidady()-VY)
  X=Posicion(Xi,10**hi,VXi,AX,Ni)
  vector_posicionx.append(X)
  vector_posicionxdeltah.append(Posicionx()-X)
  Y=Posicion(Yi,10**hi,VYi,AY,Ni)
  vector_posiciony.append(Y)
  vector_posicionydeltah.append(Posiciony()-Y)




print("Graficas en el tiempo igual a", T, "segundos:")

#Impresión de graficas
plt.plot(hrange,vector_velocidadx,'#CC0F87', label="velocidad en x")
plt.grid(color="k", linestyle="-", linewidth=.2)
plt.title("velocidad en x vs la potencia de h")
plt.xlabel("h")
plt.ylabel("velocidad")
plt.legend()
plt.show()
print("\n")

plt.plot(hrange,vector_velocidadxdeltah,'#0000CC', label="velocidad en x")
plt.grid(color="k", linestyle="-", linewidth=.2)
plt.title("$\Delta h $ en vx vs la potencia de h")
plt.xlabel("h")
plt.ylabel("$\Delta h$ en vx")
plt.legend()
plt.show()
print("\n")

plt.plot(hrange,vector_velocidady,'#CC0F87', label="velocidad en y")
plt.grid(color="k", linestyle="-", linewidth=.2)
plt.title("velocidad en y vs la potencia de h")
plt.xlabel("h")
plt.ylabel("velocidad")
plt.legend()
plt.show()
print("\n")

plt.plot(hrange,vector_velocidadydeltah,'#0000CC', label="velocidad en y")
plt.grid(color="k", linestyle="-", linewidth=.2)
plt.title("$\Delta h $ en vy vs la potencia de h")
plt.xlabel("h")
plt.ylabel("$\Delta h$ en vy")
plt.legend()
plt.show()
print("\n")

plt.plot(hrange,vector_posicionx,'#CC0F87', label="posición en x")
plt.grid(color="k", linestyle="-", linewidth=.2)
plt.title("posición en x vs la potencia de h")
plt.xlabel("h")
plt.ylabel("posición")
plt.legend()
plt.show()
print("\n")

plt.plot(hrange,vector_posicionxdeltah,'#0000CC', label="posición en x")
plt.grid(color="k", linestyle="-", linewidth=.2)
plt.title("$\Delta h $ en x vs la potencia de h")
plt.xlabel("h")
plt.ylabel("$\Delta h$ en x")
plt.legend()
plt.show()
print("\n")


plt.plot(hrange,vector_posiciony,'#CC0F87', label="posición en y")
plt.grid(color="k", linestyle="-", linewidth=.2)
plt.title("posición en y vs la potencia de h")
plt.xlabel("h")
plt.ylabel("posición")
plt.legend()
plt.show()
print("\n")

plt.plot(hrange,vector_posicionydeltah,'#0000CC', label="posición en y")
plt.grid(color="k", linestyle="-", linewidth=.2)
plt.title("$\Delta h $ en y vs la potencia de h")
plt.xlabel("h")
plt.ylabel("$\Delta h$ en y")
plt.legend()
plt.show()
print("\n")

```
